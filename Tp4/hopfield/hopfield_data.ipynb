```python
from hopfield import Hopfield, displayPatterns, exportPatterns
from letters import * #16x16 letters from a ttf file
from letters5x5 import * #5x5 letters by us
```

```python
patterns = [l,r,t,x]
model = Hopfield(size=5)
model.train(patterns)
epochs = 1
noiseLevel = 0.1
displayPatterns(patterns,["Letra C","Letra J","Letra X","Letra Z"])
```

```python
noisyC = model.addNoise(l)
resultC = model.recall(noisyC, epochs)
exportPatterns(noisyC, resultC, title="Letter C. noise level = 0.2")
noisyC = model.addNoise(l,noiseLevel)
resultC = model.recall(noisyC, epochs)
exportPatterns(noisyC, resultC, title="Letter C. noise level = 0.1")
```

```python
noisyJ = model.addNoise(r)
resultJ = model.recall(noisyJ)
exportPatterns(noisyJ, resultJ, title="Letter J. noise level = 0.2")
noisyJ = model.addNoise(r,noiseLevel)
resultJ = model.recall(noisyJ)
exportPatterns(noisyJ, resultJ, title="Letter J. noise level = 0.1")
```

```python
noisyx = model.addNoise(t)
resultx = model.recall(noisyx)
exportPatterns(noisyx, resultx, title="Letter X. noise level = 0.2")
noisyx = model.addNoise(t,noiseLevel)
resultx = model.recall(noisyx)
exportPatterns(noisyx, resultx, title="Letter X. noise level = 0.1")
```

```python
noisyz = model.addNoise(x)
resultz = model.recall(noisyz)
exportPatterns(noisyz, resultz, title="Letter Z. noise level = 0.2")
noisyz = model.addNoise(x,noiseLevel)
resultZ = model.recall(noisyz)
exportPatterns(noisyz, resultz, title="Letter Z. noise level = 0.1")
```
