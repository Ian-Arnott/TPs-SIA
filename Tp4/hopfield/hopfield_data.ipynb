```python
from hopfield import Hopfield, displayPatterns, exportPatterns
from letters import * #16x16 letters from a ttf file
from letters5x5 import * #5x5 letters by us
```

```python
patterns = [l,r,t,x]
model = Hopfield(size=5)
model.train(patterns)
epochs = 1
noiseLevel = 0.15
displayPatterns(patterns,["Letra L","Letra R","Letra T","Letra X"])
model.noiseEffectOnClassification(l, runsPerNoise=500)
model.noiseEffectOnClassification(r, runsPerNoise=500)
```

```python
noisyC = model.addNoise(l)
resultC = model.recall(noisyC, epochs)
# exportPatterns(noisyC, resultC, title="Letter C. noise level = 0.2")
model.plotRecallProgression(noisyC,maxEpochs=10,tolerance=1e-6, title="Recalls para la letra L. Noise level = 0.3")
model.plotEnergyVsEpoch(l,10,1e-6,0.4,"Energia vs Epoch. noise level = 0.5")
noisyC = model.addNoise(l,noiseLevel)
resultC = model.recall(noisyC, epochs)
# exportPatterns(noisyC, resultC, title="Letter C. noise level = 0.1")
model.plotRecallProgression(noisyC,maxEpochs=10,tolerance=1e-6, title="Recalls para la letra L. Noise level = 0.15")
```

```python
noisyJ = model.addNoise(r)
resultJ = model.recall(noisyJ)
# exportPatterns(noisyJ, resultJ, title="Letter J. noise level = 0.2")
model.plotRecallProgression(noisyJ,maxEpochs=10,tolerance=1e-6, title="Recalls para la letra R. Noise level = 0.3")
noisyJ = model.addNoise(r,noiseLevel)
resultJ = model.recall(noisyJ)
# exportPatterns(noisyJ, resultJ, title="Letter J. noise level = 0.1")
model.plotRecallProgression(noisyJ,maxEpochs=10,tolerance=1e-6, title="Recalls para la letra R. Noise level = 0.15")
```

```python
noisyx = model.addNoise(t)
resultx = model.recall(noisyx)
# exportPatterns(noisyx, resultx, title="Letter X. noise level = 0.2")
model.plotRecallProgression(noisyx,maxEpochs=10,tolerance=1e-6, title="Recalls para la letra T. Noise level = 0.3")
noisyx = model.addNoise(t,noiseLevel)
resultx = model.recall(noisyx)
# exportPatterns(noisyx, resultx, title="Letter X. noise level = 0.1")
model.plotRecallProgression(noisyx,maxEpochs=10,tolerance=1e-6, title="Recalls para la letra T. Noise level = 0.15")
```
